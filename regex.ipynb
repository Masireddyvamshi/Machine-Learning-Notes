# () groups

```python
text='cat rat cat'
re.findall('(cat)',text)
```

```python
text='cat dog cat bat rat'
re.findall(r'(cat|rat)',text)
```

```python
text = """
Contact us at:
john...doe@gmail.com
admin@yahoo.com
support@outlook...com
info@rediffmail.com
team@icloud.com
user123@gmail.com
test_user@yahoo.co.in
"""
re.findall(r'([\w\.]+)@([\w\.*]+)',text)
```

```python
text='cat rat catcatcat'
re.findall('(cat)+',text)
# returns last captured group
```

```python
mo=re.search('(bat)*(ball)','batball batter baller')
mo
```

```python
mo.group()
```

```python
mo.group(1)
```

```python
mo.group(2)
```

```python
mo = re.search(r'(?P<G1>(bat)*(ball))', 'batball batter baller')
```

```python
mo.group()
```

```python
mo.group('G1')
```

```python
mo.groups()
```
