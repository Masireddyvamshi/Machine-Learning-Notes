```python
pd.read_csv(r"C:\Users\masir\Downloads\Machine Learning\Handling_different_formats\Encoding_error\die_ISO-8859-1.txt")
```

```python
encodings_list = [
    "utf-8",
    "utf-8-sig",
    "ISO-8859-1",
    "latin1",
    "cp1252",
    "utf-16"
] 
```

```python
for i in encodings_list:
    try:
        df=pd.read_csv(r"C:\Users\masir\Downloads\Machine Learning\Handling_different_formats\Encoding_error\die_ISO-8859-1.txt",encoding=i,on_bad_lines='skip')
    except UnicodeDecodeError:
        continue
```

```python
df
```

```python
print(f'length: {len(encodings.aliases.aliases)}') #dictionary of encodings 
print(encodings.aliases.aliases)
```

```python
encodings_list=encodings.aliases.aliases.keys()
# keys are acting as encodings
```

```python
suitable_encodings = []
not_suitable_encodings = []
unsupported = []

for i in encodings_list:
    try:
        df = pd.read_csv(
            r"C:\Users\masir\Downloads\Machine Learning\Handling_different_formats\Encoding_error\file_guide.csv",
            encoding=i,
            on_bad_lines='skip'
        )
        suitable_encodings.append(i)

    except UnicodeError:
        # All decoding/translation failures
        not_suitable_encodings.append(i)

    except LookupError:
        # Encoding name not recognized
        unsupported.append(i)
```

```python
print(len(suitable_encodings),len(not_suitable_encodings),len(unsupported))
unsupported
```
