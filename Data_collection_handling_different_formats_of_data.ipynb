```python
print(f'length: {len(encodings.aliases.aliases)}') #dictionary of encodings 
print(encodings.aliases.aliases)
```

```python
encodings_list=encodings.aliases.aliases.keys()
# keys are acting as encodings
```

```python
suitable_encodings = []
not_suitable_encodings = []
unsupported = []

for i in encodings_list:
    try:
        df = pd.read_csv(
            r"C:\Users\masir\Downloads\Machine Learning\Encoding_error\file_guide.csv",
            encoding=i,
            on_bad_lines='skip'
        )
        suitable_encodings.append(i)

    except UnicodeError:
        # All decoding/translation failures
        not_suitable_encodings.append(i)

    except LookupError:
        # Encoding name not recognized
        unsupported.append(i)
```

```python
print(len(suitable_encodings),len(not_suitable_encodings),len(unsupported))
unsupported
```

```python
data = {
    "order_id": 101,
    "customer": {
        "customer_id": "C001",
        "name": "Rahul",
        "contact": {
            "email": "rahul@gmail.com",
            "phone": "9876543210"
        }
    },
    "items": [
        {
            "product_id": "P001",
            "product_name": "Laptop",
            "price": 60000,
            "quantity": 1
        },
        {
            "product_id": "P002",
            "product_name": "Mouse",
            "price": 800,
            "quantity": 2
        }
    ],
    "payment": {
        "method": "Credit Card",
        "status": "Success"
    }
}
```

```python
pd.json_normalize(data,record_path=['items']) 
```

```python
pd.json_normalize(data,record_path=['items'],meta=['order_id','customer','payment']) 
```

```python
pd.json_normalize(data,record_path=['items'],meta=['order_id',['customer','customer_id'],['customer','name'],['customer','contact'],'payment']) 
```

```python
pd.json_normalize(data,record_path=['items'],meta=['order_id',['customer','customer_id'],['customer','name'],['customer','contact'],['customer','contact','email'],['customer','contact','phone'],'payment']) 
```
