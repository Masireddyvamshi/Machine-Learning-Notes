```python
import re
```

# re.search()

```python
text='hello sir my name is vamshidhar, how is your day sir'
re.search('sir',text)    
```

```python
text='hello sir my name is vamshidhar, how is your day sir'
if re.search('sir',text):
    print('Match Found')
else:
    print('Match Not Found')
```

# character class ([abc])

```python
pattern=r'[b12c]'
re.search(pattern,'baaa')
```

```python
pattern=r'[b12c]'
re.search(pattern,'aca1a')
```

# re.findall()

```python
pattern=r'[b12c]'
re.findall(pattern,'aa1cdb')
```

```python
text='i love biryani at pistahouse'
re.findall('[a-z][a-z]',text)
```

```python
text='i love biryani at pistahouse'
re.findall('[a-z][a-z][a-z]',text)
```

```python
text='i 143 biryani 245 at pistahouse 987'
re.findall('[0-9][0-9][0-9]',text)
```

```python
text='i 143 biryani 2@5 at pistahouse 987'
re.findall('[0-9][@][0-9]',text)
```

```python
text='i Love To 2cd #2de 2#d'
re.findall('[a-zA-Z0-9][a-zA-Z0-9]',text)
```

# dot character(.)

```python
text='I Love India '
re.findall('.',text)
```

```python
text='I Love India #proud to be an Iindian143'
re.findall('..',text)
```

```python
text='I Love India #proud to be an indian\n143'
re.findall('...',text)
```

# \w and \W

```python
text='i Love to\n2cd_'
re.findall('\w',text)
```

```python
text='i Love to\n2cd_ __'
re.findall('\w\w',text)
```

```python
text='i Love to\n2cd_'
re.findall('\w\w\w',text)
```

```python
text='I Love to\n2cd_ __'
re.findall('\W',text)
```

```python
text='I Love to\n 2cd_ __'
re.findall('\W\W',text)
```

```python
text='I Love to\n 2cd_ __'
re.findall('\w\W',text)
```

```python
text='I Love to\n 2cd_ __'
re.findall('\W\w',text)
```

```python
text='I Love to\n 2cd_ __'
re.findall('.\W\w',text) 
```

# \d and \D

```python
text='I Love 143 India 645'
re.findall('\d',text)
```

```python
text='I Love 143 India 645'
re.findall('\d\d\d',text)
```

```python
text='I Love 143 India 645'
re.findall('\D',text)
```
