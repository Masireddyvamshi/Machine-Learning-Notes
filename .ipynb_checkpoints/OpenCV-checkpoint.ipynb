# Live Webcam Videocapture 

```python
vid=cv2.VideoCapture(0)
while True:
    success,frame=vid.read()
    if success:
        cv2.imshow('Live Stream',frame)
    else:
        break
    if cv2.waitKey(1) & 255 == ord('q'):
        break
vid.release()
cv2.destroyAllWindows()
```

```python
vid=cv2.VideoCapture(0)
while True:
    success,frame=vid.read()
    if success:
        shape=frame.shape[:-1]
        b,g,r=cv2.split(frame)
        zeros=np.zeros(shape,dtype=np.uint8)
        blue_channel=cv2.merge([b,zeros,zeros])
        green_channel=cv2.merge([zeros,g,zeros])
        red_channel=cv2.merge([zeros,zeros,r])
        gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        cv2.imshow('Live Stream',frame)
        cv2.imshow('gray Live Stream',gray)
        cv2.imshow('Blue_channel',blue_channel)
        cv2.imshow('green_channel',green_channel)
        cv2.imshow('red_channel',red_channel)
    else:
        break
    if cv2.waitKey(1) & 255 == ord('q'):
        break
vid.release()
cv2.destroyAllWindows()
```

# scaling

```python
img=cv2.imread(r"C:\Users\masir\Downloads\Machine Learning\OpenCV\dog.jpg")
img=cv2.resize(img,None,fx=0.1,fy=0.1)
```

```python
sx=1
sy=1
tx=0
ty=0
scaled_Matrix=np.array([[sx,0,tx],
                       [0,sy,ty]],dtype=np.float32)
h,w=img.shape[:2][::-1]
Scaled_image=cv2.warpAffine(img,scaled_Matrix,(sx*h,sy*w))
cv2.imshow('OrginalImage',img)
cv2.imshow('scaled Image',Scaled_image)
cv2.waitKey()
cv2.destroyAllWindows()
```
